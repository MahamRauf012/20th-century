```python
# Core
import re, string, collections, json, os
from collections import Counter

# Data / viz
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# NLP
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from textblob import TextBlob

# one-time NLTK downloads (safe to run twice)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')

sns.set()  # seaborn default theme
plt.rcParams['figure.figsize'] = (10,5)
```

```python
with open(r"C:\Users\maham\20th-century\20th_century_text.txt", "r", encoding="utf-8", errors="ignore") as file:
    raw_text = file.read().replace('\n', ' ')
print(raw_text[:500])
```

```python
import nltk
nltk.download('punkt')                    # tokenizers
nltk.download('punkt_tab')                # NEW in recent NLTK
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
```

```python
tokens = word_tokenize(raw_text)
fdist = nltk.FreqDist(tokens)
top10_raw = pd.DataFrame(fdist.most_common(10), columns=["token", "count"])

sns.barplot(data=top10_raw, x="token", y="count")
plt.title("Top 10 tokens (raw)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
import string
from nltk.corpus import stopwords

# Step 1: Define stopwords and punctuation
stop_words = set(stopwords.words('english'))
punct = set(string.punctuation)

# Step 2: Clean tokens (remove punctuation, lowercase, remove stopwords)
clean_tokens = [
    word.lower()
    for word in tokens
    if word.isalpha() and word.lower() not in stop_words and word not in punct
]

# Step 3: Frequency distribution of cleaned text
clean_fdist = nltk.FreqDist(clean_tokens)

# Step 4: Top 10 cleaned tokens
top10_clean = pd.DataFrame(clean_fdist.most_common(10), columns=["token", "count"])

# Step 5: Visualize
sns.barplot(data=top10_clean, x="token", y="count")
plt.title("Top 10 Tokens (Cleaned Text)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
from nltk import pos_tag
pos_tags = pos_tag(clean_tokens)
pos_df = pd.DataFrame(pos_tags, columns=['word', 'pos'])
```

```python
# Filter for nouns, verbs, and adjectives
nouns = pos_df[pos_df['pos'].str.startswith('NN')]
verbs = pos_df[pos_df['pos'].str.startswith('VB')]
adjectives = pos_df[pos_df['pos'].str.startswith('JJ')]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Top 15 nouns
top_nouns = nouns['word'].value_counts().head(15)
plt.figure(figsize=(10,5))
sns.barplot(x=top_nouns.values, y=top_nouns.index, palette='viridis')
plt.title("Top 15 Nouns in 20th-Century Text")
plt.xlabel("Frequency")
plt.ylabel("Nouns")
plt.show()

# Top 15 verbs
top_verbs = verbs['word'].value_counts().head(15)
plt.figure(figsize=(10,5))
sns.barplot(x=top_verbs.values, y=top_verbs.index, palette='cool')
plt.title("Top 15 Verbs in 20th-Century Text")
plt.xlabel("Frequency")
plt.ylabel("Verbs")
plt.show()

# Top 15 adjectives
top_adjs = adjectives['word'].value_counts().head(15)
plt.figure(figsize=(10,5))
sns.barplot(x=top_adjs.values, y=top_adjs.index, palette='magma')
plt.title("Top 15 Adjectives in 20th-Century Text")
plt.xlabel("Frequency")
plt.ylabel("Adjectives")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define a lookup list of countries
countries = [
    "United States", "USA", "Russia", "Soviet Union", "China", "Germany",
    "France", "United Kingdom", "Britain", "Japan", "India", "Italy", "Canada",
    "Australia", "Spain", "Poland"
]

# Create a dictionary to store counts
country_counts = {}

# Count mentions (case-insensitive)
for country in countries:
    country_counts[country] = raw_text.lower().count(country.lower())

# Convert to DataFrame
country_df = pd.DataFrame(list(country_counts.items()), columns=["Country", "Mentions"])

# Sort for plotting
country_df = country_df.sort_values(by="Mentions", ascending=False)

# Display DataFrame
display(country_df)

# Plot country mention frequencies
plt.figure(figsize=(10,6))
sns.barplot(x="Mentions", y="Country", data=country_df, palette="crest")
plt.title("Frequency of Country Mentions in 20th-Century Text")
plt.xlabel("Number of Mentions")
plt.ylabel("Country")
plt.show()
```

```python
# Plot frequency of country mentions
plt.figure(figsize=(10,6))
sns.barplot(x="Mentions", y="Country", data=country_df, palette="viridis")
plt.title("Country Mentions in the 20th-Century Text", fontsize=14)
plt.xlabel("Number of Mentions", fontsize=12)
plt.ylabel("Country", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
Findings:
The bar chart shows that certain countries  particularly the United States, Soviet Union, and Germany  are mentioned far more often than others. This pattern aligns with the major global events of the twentieth century such as the World Wars, the Cold War, and the emergence of the U.S. and U.S.S.R. as superpowers.

The moderate mentions of China, Japan, and Britain reflect their involvement in regional conflicts and economic transformations, while countries like India, Spain, and Poland appear less frequently, possibly due to their more localized roles in historical narratives.

Overall, the distribution highlights how geopolitical power and global influence shaped the dominant themes in twentieth-century history.
```
